intermezzo : coding style
PEP 8 has emerged as the style guide that promotes a very readable and eye pleasing coding style. 
use 4 space indentations and no tabs
wrap lines so that they do not exceed 79 characters
use blank lines to separate functions and classes. and larger blcoks of code inside functions
when possible, put comments on a line of their own
use docstrings
use spaces around operators and after commas, but not directly insdie bracketing constructs
a = f(1,2) + g(3,4)

name your classes and functions consistently, convention is to use UpperCamelCase for classes and lower_with_underscores for functions and methods
   
always use self as the name for the first method argument
don't use fancy codings for your code
